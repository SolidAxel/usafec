```python
#df = df.drop(['cmte_id', 'amount'], axis='columns')
df
```

```python
#different data will be printed, by changing year number
df1982 = df.loc[df['year'] == 2020]
df1981 = df.loc[df['year'] == 2019]
```

```python
statelist = ['NY', 'AZ', 'CA', 'NV', 'AK', 'HI', 'OR', 'WA', 'ID', 'MT', 'UT', 'WY','CO','NM','TX', 'OK', 'KS',
            'NE','SD','ND','WI','IA','MN','IL','MO','AR','LA','MS','AL','TN','KY','IN','OH','MI','FL','GA',
            'SC', 'NC','VA','WV','MD', 'DE','PA','NJ','CT','RI','MA','VT','NH','MI']
demsumlist=[]
repsumlist=[]
indsumlist=[]
```

```python
for i in statelist:
    asd=df1982.loc[df1982['party']=='DEM'].loc[df1982['state']==i]
    sss=asd['adjusted']
    sum1=sum(sss)
    asd=df1981.loc[df1981['party']=='DEM'].loc[df1981['state']==i]
    ssss=asd['adjusted']
    sum2=sum(ssss)
    value=sum1+sum2
    demsumlist.append(value)
    asd=df1982.loc[df1982['party']=='REP'].loc[df1982['state']==i]
    sss=asd['adjusted']
    sum1=sum(sss)
    asd=df1981.loc[df1981['party']=='REP'].loc[df1981['state']==i]
    ssss=asd['adjusted']
    sum2=sum(ssss)
    value=sum1+sum2
    repsumlist.append(value)
```

```python
repsumlist
```

```python
for (i,j,k) in zip(statelist, repsumlist, demsumlist):
    print("{ state: '"+i+"', rep: "+str(j)+", dem: "+str(k)+" },")
```
